```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data =pd.read_csv('survey lung cancer.csv')
```

```python
print(Data)
```

```python
print('In this data 2 indicates as yes 1 as no')
```

```python
print(sns.heatmap(Data.corr()))
```

```python
Data['GENDER']=pd.factorize(Data['GENDER'])[0]
Data['LUNG_CANCER']=pd.factorize(Data['LUNG_CANCER'])[0]
```

```python
print(Data)
```

```python
print('After factorizing gender M refers to 0 and F refers to 1')
print('After factorizing gender YES refers to 0 and NO refers to 1')
```

```python
X=Data.values[:,:-1]

print("The value shows data",X)
Y=Data.values[:,15:16]

print("The value shows result of data wether Cancer detected or not",Y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB

GNB=GaussianNB()
```

```python
GNB.fit(X_train,Y_train)
```

```python
pred=GNB.predict(X_test)
```

```python
print(pred)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(Y_test,pred)
```

```python
print('Model prediction is', score*100)
```
